The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

    7. Create a spending flag

```python
#Split the data into groups based on user id; apply transform on the order number column; create a new column
ords_prods_merge['ave_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Check the output
ords_prods_merge.head(20)
```

```python
#Create a flag for average spend
ords_prods_merge.loc[ords_prods_merge['ave_spend'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['ave_spend'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#Review Output
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

    8. Create an order frequency flag

```python
#Split the data into groups based on user id; apply transform on the days since prior order column; create a new column
ords_prods_merge['median_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Review the output
ords_prods_merge.head(40)
```

The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#Create a flag for order frequency label
ords_prods_merge.loc[ords_prods_merge['median_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order'] <=20) & (ords_prods_merge['median_order'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order'] < 10, 'frequency_flag'] = 'Frequent customer' 
```

```python
#Check the frequency
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```

```python
#Missing the equal sign in frequent customer
ords_prods_merge.loc[ords_prods_merge['median_order'] <= 10, 'frequency_flag'] = 'frequent customer'
```

```python
#recheck the frequency
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
